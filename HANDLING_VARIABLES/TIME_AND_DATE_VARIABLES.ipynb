```python
#from above we see the date Dtype is object we need to convert it to Datetime
date['date']=pd.to_datetime(date['date'])
```

## EXTRACT YEAR

```python
date['year']=date['date'].dt.year
```

```python
date.head()
```

## Extract day

```python
date['day']=date['date'].dt.day
```

```python
date.head()
```

## Extract Month

```python
date['month']=date['date'].dt.month
```

```python
date.head()
```

## Extract Month name 

```python
date['month_name']=date['date'].dt.month_name()
```

```python
date.head()
```

## Extract day of week

```python
date['day of week']=date['date'].dt.dayofweek
```

```python
date.head()
```

## Extract name of week

```python
date['day_name']=date['date'].dt.day_name()
```

```python
date.head()
```

```python
# is weekend?
date['weekend']=np.where(date['day_name'].isin(['Sunday','Saturday']),1,0)
```

```python
date.head()
```

## Extract week in a year

```python
date['week_in_year']=date['date'].dt.week
```
